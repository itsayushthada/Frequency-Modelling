# Visualization

## Distance Frequency Plots

```python
def DFgraph(depth):

    idx = df.d == depth

    plt.rcParams["figure.figsize"] = 16, 8
    print("Distance-Frequency Plot for Crack of {} Depth".format(depth))
    print("-------------------------------------------------\n".format(depth))
    
    plt.subplot(2, 3, 1)
    plt.plot(df[idx].l1, df[idx].w1, "r")
    plt.xlabel("Distance of I crack.")
    plt.ylabel("I Natural Frequency")

    plt.subplot(2, 3, 2)
    plt.plot(df[idx].l1, df[idx].w2, "g")
    plt.xlabel("Distance of I crack.")
    plt.ylabel("II Natural Frequency")

    plt.subplot(2, 3, 3)
    plt.plot(df[idx].l1, df[idx].w3, "b")
    plt.xlabel("Distance of I crack.")
    plt.ylabel("III Natural Frequency")

    plt.subplot(2, 3, 4)
    plt.plot(df[idx].l2, df[idx].w1, "r")
    plt.xlabel("Distance of II crack.")
    plt.ylabel("I Natural Frequency")

    plt.subplot(2, 3, 5)
    plt.plot(df[idx].l2, df[idx].w2, "g")
    plt.xlabel("Distance of II crack.")
    plt.ylabel("II Natural Frequency")

    plt.subplot(2, 3, 6)
    plt.plot(df[idx].l2, df[idx].w3, "b")
    plt.xlabel("Distance of II crack.")
    plt.ylabel("III Natural Frequency")

    plt.show()
```

## Distance Frequency Plots with Multiple Crack Distances

```python
def DFgraph_multiL(depth):

    idx = df.d == depth

    plt.rcParams["figure.figsize"] = 16, 5
    print("Distance-Frequency Plot for Crack of {} Depth".format(depth))
    print("-------------------------------------------------\n".format(depth))
    
    plt.subplot(1,3,1)
    plt.plot(df[idx].l1, df[idx].w1)
    plt.plot(df[idx].l2, df[idx].w1)
    plt.xlabel("Distance of crack.")
    plt.ylabel("I Natural Frequency")

    plt.subplot(1,3,2)
    plt.plot(df[idx].l1, df[idx].w2)
    plt.plot(df[idx].l2, df[idx].w2)
    plt.xlabel("Distance of crack.")
    plt.ylabel("II Natural Frequency")

    plt.subplot(1,3,3)
    plt.plot(df[idx].l1, df[idx].w3)
    plt.plot(df[idx].l2, df[idx].w3)
    plt.xlabel("Distance of crack.")
    plt.ylabel("III Natural Frequency")

    plt.show()
```

```python
DFgraph_multiL(0.0020)
```

```python
DFgraph_multiL(0.0021)
```

```python
DFgraph_multiL(0.0022)
```

```python
DFgraph_multiL(0.0023)
```

```python
DFgraph_multiL(0.0024)
```

```python
DFgraph_multiL(0.0025)
```
