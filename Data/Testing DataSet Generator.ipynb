This notebook is meant to merge the data table which contains the details of the experiments and the table which contains the results scrapped from the reports geenrated by Ansys.<br>

The process mentioned above is known as Table Join.

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("./test_metadata-free-free-system.csv")
df1.head()
```

```python
df2 = pd.read_csv("../Reports/Fixed-Fixed-Scrapped-Data/Fixed-Fixed-Test.csv")
df2.head(18)
```

# Transformation of Tables

```python
# Transformation on Table I

df1[["l1", "l2"]] = df1[["l1", "l2"]].astype(int)
```

```python
# Transformation on Table II

df2[["Index", "Mode"]] = df2[["Index", "Mode"]].astype(int)
df2[["Index"]] -= 1
```

# Joining of Table

```python
result = df1.join(df2.set_index("Index"))
result.head(18)
```

```python
# Removing the Spurious Recors

result = result[~np.isnan(result["Mode"])]
```

```python
result["Mode"] = result["Mode"].astype(int)
```

# Exporting the Table

```python
result.to_csv("Fixed-Fixed-Test.csv", index=False)
result.to_excel("Fixed-Fixed-Test.xlsx", index=False)
```
