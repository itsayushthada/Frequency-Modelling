```python
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_exp1 = pd.read_excel("../Result.xlsx", sheet_name="csv_Method1")
df_exp1
```

```python
df_exp2 = pd.read_excel("../Result.xlsx", sheet_name="csv_Method2")
df_exp2
```

```python
alpha = 0.05

for col in df_exp1.columns[1:]:
    statistic, pval = stats.mannwhitneyu(df_exp1[col], df_exp2[col], alternative=None)
    print("Column: {}  Statistics: {:07.3f}   pval: {:.15f}  Null_Rejected: {}".format(col.ljust(9), 
                                                                                       statistic, 
                                                                                       pval, 
                                                                                       pval<alpha))
```
